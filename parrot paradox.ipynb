```python
def time_switch(price_A,price_B,capital,T,t1,L,alpha):
    n_a = 0
    n_b = 0
    cap_hist = []
    for i in range(0,L):
        if i%T == 0:
            #sell stock B
            capital = capital + (1-alpha)*(n_b)*price_B[i]
            n_b = 0
            cap_hist.append(capital)
            #buy stock A
            n_a = (capital*(1-alpha))//price_A[i]
            capital = capital - (n_a*price_A[i])
            
        elif i%T == t1:
            
            #sell stock A
            capital = capital + (1-alpha)*(n_a)*price_A[i]
            n_a = 0
            cap_hist.append(capital)
            #buy stock B
            n_b = (capital*(1-alpha))//price_B[i]
            capital = capital - (n_b*price_B[i])
        else:
            #check net worth
            temp = capital + (1-alpha)*(n_b)*price_B[i] + (1-alpha)*(n_a)*price_A[i]
            cap_hist.append(temp)
            

    
    return cap_hist 


#set stock A and B
label_a = "LABD"
label_b = "SPXU"
price_a = price_SDOW
price_b = price_ery
L = len(price_smdd)


#print(len(price_xrp))
#print(len(price_ada))



T = 50
t1 = 32


plt.plot(price_a, label=label_a)
plt.plot(price_b, label=label_b)
plt.legend([label_a, label_b])



# capital:10000 alpha(trans fee) : 0.01
hist = time_switch(price_a,price_b,10000,T,t1,L,0.001)
print(hist[-1])

ls = []
for i in range(T):
    temp = time_switch(price_a,price_b,10000,T,i,L,0.001)
    ls.append(round(temp[-1],1))
print(ls)
print(T)
```

```python
def high_pass_filter(sig,time_vec,peak):
    time_step = 0.02
    period = 5.

    # The FFT of the signal
    sig_fft = fftpack.fft(sig)

    # And the power (sig_fft is of complex dtype)
    power = np.abs(sig_fft)**2

    # The corresponding frequencies
    sample_freq = fftpack.fftfreq(sig.size, d=time_step)



    # Find the peak frequency: we can focus on only the positive frequencies
    pos_mask = np.where(sample_freq > 0)
    freqs = sample_freq[pos_mask]
    peak_freq = freqs[power[pos_mask].argmax()]

    # Check that it does indeed correspond to the frequency that we generate
    # the signal with
    np.allclose(peak_freq, 1./period)



    # scipy.signal.find_peaks_cwt can also be used for more advanced
    # peak detection

    high_freq_fft = sig_fft.copy()
    high_freq_fft[np.abs(sample_freq) > peak] = 0
    filtered_sig = fftpack.ifft(high_freq_fft)
    
    plt.figure(figsize=(6, 5))
    plt.plot(time_vec, sig, label='Original signal')
    plt.plot(time_vec, filtered_sig, linewidth=3, label='Filtered signal')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')

    plt.legend(loc='best')
    return filtered_sig
    
clean_a = high_pass_filter(price_a,time,5)
clean_b = high_pass_filter(price_b,time,5)
# capital:10000 alpha(trans fee) : 0.01
clean_hist = time_switch(clean_a,clean_b,10000,T,t1,L,0.001)
print(clean_hist[-1])

ls = []
for i in range(T):
    temp = time_switch(clean_a,clean_b,10000,T,i,L,0.001)
    ls.append(round(abs(temp[-1]),1))
print(ls)
print(T)
```
